## Balancing dataset

### Utilizing Random undersampling
This can be used to work on imbalance dataset, This method seeks to randomly select and remove samples from the majority class, consequently reducing the number of examples in the majority class in the transformed data.

```python
from imblearn.under_sampling import RandomUnderSampler, NearMiss
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler, SMOTE
from sklearn.model_selection import RepeatedStratifiedKFold
from imblearn.pipeline import Pipeline
```

```python
rand_under= RandomUnderSampler(sampling_strategy='majority', random_state=12)  

X_und, y_und= rand_under.fit_resample(X, y)

ax= y_und.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title('Undersampling')

# checking to see the number of records after over sampling
print(sorted(Counter(y_und).items()))
```

```python
# Create Decision Tree classifer object
model_und = CatBoostClassifier(silent= True)

# Train Decision Tree Classifer
model_und = model_und.fit(X_und, y_und)

#Predict the response for test dataset
y_pred_und = model_und.predict(X_test)

print("Accuracy:", metrics.accuracy_score(y_test, y_pred_und))

# check model performance
print(classification_report(y_test, y_pred_und))
```

### Performing Random oversampling to balance distribution
Random Oversampling includes selecting random examples from the minority class with replacement and supplementing the training data with multiple copies of this instance. the random oversampling may increase the likelihood of occurring overfitting, since it makes exact copies of the minority class examples. In this way, a symbolic classifier, for instance, might construct rules that are apparently accurate, but actually cover one replicated example

```python
rand_over= RandomOverSampler(sampling_strategy='minority', random_state=12)  

X_over, y_over= rand_over.fit_resample(X, y)

ax= y_over.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title('Oversampling')

# checking to see the number of records after over sampling
print(sorted(Counter(y_over).items()))
```

```python
# Create Decision Tree classifer object
model_over = CatBoostClassifier(silent= True)

# Train Decision Tree Classifer
model_over = model_over.fit(X_over, y_over)

#Predict the response for test dataset
y_pred_over = model_over.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred_over))

print(classification_report(y_test, y_pred_over))
```

### Synthetic Minority Oversampling Techinque
SMOTE first selects a minority class instance a at random and finds its k nearest minority class neighbors. The synthetic instance is then created by choosing one of the k nearest neighbors b at random and connecting a and b to form a line segment in the feature space. The synthetic instances are generated as a convex combination of the two chosen instances a and b.
https://machinelearningmastery.com/random-oversampling-and-undersampling-for-imbalanced-classification/

```python
smote= SMOTE(sampling_strategy='auto', random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

ax= y_smote.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title('SMOTE')

# checking to see the number of records after over sampling
print(sorted(Counter(y_smote).items()))
```

```python
# Create Decision Tree classifer object
model_smote = CatBoostClassifier(silent= True)

# Train Decision Tree Classifer
model_smote = model_smote.fit(X_smote, y_smote)

#Predict the response for test dataset
y_pred_smote = model_smote.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred_over))

print(classification_report(y_test, y_pred_smote))
```

### Undersampling using Nearmiss
This uses KNN model to perform undersampling.
Types of the near-miss algorithm:
- Version 1: In the first version, the data is balanced by calculating the average minimum distance between the larger distribution and three closest smaller distributions.
- Version 2: Here, the data is balanced by calculating the average minimum distance between the larger distribution and three furthest smaller distributions. 
- Version 3: Here, the smaller class instances are considered and m neighbours are stored. Then the distance between this and the larger distribution is taken and the largest distance is eliminated.
https://analyticsindiamag.com/using-near-miss-algorithm-for-imbalanced-datasets/

```python
nearmiss= NearMiss(sampling_strategy='auto', version=3)
X_nearmiss, y_nearmiss = nearmiss.fit_resample(X, y)

ax= y_nearmiss.value_counts().plot.pie(autopct='%.2f')
_ = ax.set_title('Nearmiss')

# checking to see the number of records after over sampling
print(sorted(Counter(y_nearmiss).items()))
```

```python
# Create Decision Tree classifer object
model_nearmiss = CatBoostClassifier(silent= True, seed= 12)

# Train Decision Tree Classifer
model_nearmiss = model_nearmiss.fit(X_nearmiss, y_nearmiss)

#Predict the response for test dataset
y_pred_nearmiss = model_nearmiss.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred_nearmiss))

print(classification_report(y_test, y_pred_nearmiss))
```

#### Using Bayesian optimization for XGBoost
There are different types of Bayesian optimization. Hyperopt is used in this example. 
We defined an objective function that takes in the parameters and returns the loss. Since the goal is to maximize the recall value, we set max(scores) as the best_score, and set the loss to be -best_score. This setting ensures maximizing recall while minimizing the loss. fmin is used to optimize the objective function. Hyperopt currently has three algorithms. They are random search, Tree of Parzen Estimators (TPE), and adaptive TPE. We are using TPE as the search algorithm

```python
space={'max_depth': hp.quniform("max_depth", 3, 18, 1),
        'gamma': hp.uniform ('gamma', 1,9),
        'reg_alpha' : hp.quniform('reg_alpha', 40,180,1),
        'reg_lambda' : hp.uniform('reg_lambda', 0,1),
        'colsample_bytree' : hp.uniform('colsample_bytree', 0.5,1),
        'min_child_weight' : hp.quniform('min_child_weight', 0, 10, 1),
        'n_estimators': 180,
        'seed': 0
    }
```

```python
def objective(space):
    clf= XGBClassifier(
                    n_estimators =space['n_estimators'], max_depth = int(space['max_depth']), gamma = space['gamma'],
                    reg_alpha = int(space['reg_alpha']),min_child_weight=int(space['min_child_weight']),
                    colsample_bytree=int(space['colsample_bytree']))
    
    evaluation = [( X_train, y_train), ( X_test, y_test)]
    
    clf.fit(X_train, y_train,
            eval_set=evaluation, eval_metric="auc",
            early_stopping_rounds=10, verbose=False)
    

    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, pred>0.5)
    print ("SCORE:", accuracy)
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
trials = Trials()

best_hyperparams = fmin(fn = objective,
                        space = space,
                        algo = tpe.suggest,
                        max_evals = 100,
                        trials = trials)
```

```python
print("The best hyperparameters are : ","\n")
print(best_hyperparams)
```

```python
# Train model using the best parameters
xgboost_ba = XGBClassifier(seed=0, 
                           colsample_bytree=0.8, 
                           gamma=3.66, 
                           learning_rate=0.04, 
                           max_depth=7, 
                           reg_alpha=83.0,
                           reg_lambda=0.95
                           ).fit(X_train, y_train)
```

```python
# Make prediction using the best model
bayesian_opt_predict = xgboost_ba.predict(X_test)

# Get predicted probabilities
bayesian_opt_predict_prob = xgboost_ba.predict_proba(X_test)[:,1]

# Get performance metrics
precision, recall, fscore, support = score(y_test, bayesian_opt_predict)

# Print result
print(f'The recall value for the xgboost Bayesian optimization is {recall[1]:.4f}')
```
